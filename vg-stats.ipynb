# Lab: Data Analysis with Pandas
## Video Game Stats

```python
df = pd.read_csv('/kaggle/input/videogamesales/vgsales.csv')
df
```

# Which company is the most common video game publisher?

```python
most_common_pub = df.groupby('Publisher').count().sort_values(by="Rank", ascending=False).head(1).index[0]
most_common_pub
```

# What's the most common plaform?

```python
most_common_platform = df.groupby('Platform').count().sort_values(by="Rank", ascending=False).head(1).index[0]
most_common_platform
```

# What about the most common genre?

```python
most_common_genre = df.groupby('Genre').count().sort_values(by='Rank', ascending=False).head(1).index[0]
most_common_genre
```

# What are the top 20 highest grossing games?

```python
highest_grossing = df.sort_values(by='Rank').head(20)['Name']
highest_grossing
```

# For North American video game sales, whatâ€™s the median?

- Provide a secondary output showing ten games surrounding the median sales output.
- Assume that games with same median value are sorted in descending order.

```python
north_american_median = df.NA_Sales.median()
print(north_american_median)
north_american_median_index = df.index[df.NA_Sales == north_american_median][0]
print(df.iloc[north_american_median_index-5:north_american_median_index+5]['Name'])
```

# For the top-selling game of all time, how many standard deviations above/below the mean are its sales for North America?

```python
std_dev = df.NA_Sales.std()
top_seller = df.sort_values(by='NA_Sales').head(1)['NA_Sales']
(top_seller - north_american_median) / std_dev
```

# The Nintendo Wii seems to have outdone itself with games. How does its average number of sales compare with all of the other platforms?

```python
platform_avgs = df.groupby(by='Platform')['Global_Sales'].mean().sort_values(ascending=False)
platform_avgs
```

# Are there any games that have made the Top 20 in sales in all 3 regions of North America, Europe, and Japan?

```python
top_na = df.sort_values(by='NA_Sales', ascending=False).head(20)
top_eu = df.sort_values(by='EU_Sales', ascending=False).head(20)
top_jp = df.sort_values(by='JP_Sales', ascending=False).head(20)
top_all = pd.merge(top_na, top_eu, on='Name').merge(top_jp, on='Name')['Name']
top_all
```

# What year(s) was/were the real 'heyday' of video games, i.e., what year of release had the most titles finish in the Top 300 for global sales?

```python
best_year = df.sort_values(by='Global_Sales', ascending=False).head(300)['Year'].mode()
best_year
```

# Who is the publisher that has sold the most video games globally each year?

```python
yearly_sales_by_pub = df.groupby(['Year', 'Publisher'])['Global_Sales'].sum()
idx = yearly_sales_by_pub.groupby('Year').idxmax()
yearly_sales_by_pub.loc[idx]
```
